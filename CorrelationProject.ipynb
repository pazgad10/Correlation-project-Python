```python
# import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Calling the data
df = pd.read_csv(r'C:\Users\paz\Downloads\movies.csv')
```

```python
# Looking for missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,percent_missing))
```

```python
# Data types for our columns

print(df.dtypes)
```

```python
# Changing data type of columns for arraging them

df['votes'] = df['votes'].fillna(0)
df['votes'] = df['votes'].astype('int64')

df['budget'] = df['budget'].fillna(0)
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].fillna(0)
df['gross'] = df['gross'].astype('int64')

df['runtime'] = df['runtime'].fillna(0)
df['runtime'] = df['runtime'].astype('int64')
```

```python
df
```

```python
#Making sure the year is correct by making a new correct year column

df['YearCorrect'] = df['released'].str.rsplit(',').str[-1].str[:5]
df
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget For Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Top 10 companies by year and gross revenue

CompanyGrossSum = df.groupby(['company', 'YearCorrect'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross', 'company', 'YearCorrect'], ascending = False)[:10]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Up here we can see high correlation between budget and gross, and same with votes and gross.

# Let's try an extensive correlation with all the categories and see if the result will change.

# I think that when we will go deeper we will see that the budget and gross has high correlation because the budget inculudes important categories like stars.
```

```python
# Assining a random numeric value for each category using factorize

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix2 = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix2, annot=True)

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_matrix3 = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_matrix3.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values()  

print(sorted_pairs)
```

```python
pd.set_option('display.max_rows', None)
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

print(high_corr)

```
